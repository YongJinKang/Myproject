### Feature Engineering 

### Fix the typo
: handcap이라고 써져 있는 오타를 handicap으로 수정

```python
df.rename(columns={"Handcap": "Hadicap"}, inplace = True) 
                   # inplace= True로 넣으면 굳이 df에 넣지 않아도 된다.
```

### Change the data type
: 날짜와 시간(00:00:00)까지 모두 나와있는 AppointmentDay와 

```python
df.AppointmentDay = pd.to_datetime(df.AppointmentDay).dt.date
df.ScheduledDay = pd.to_datetime(df.ScheduledDay).dt.date
print(df.AppointmentDay,df.ScheduledDay)
```

### 기간 구하기 : 파생변수 만들기

```python
df['PeriodBetween'] = df.AppointmentDay - df.ScheduledDay
df['PeriodBetween'] = df['PeriodBetween'].dt.days
df.head()
```
